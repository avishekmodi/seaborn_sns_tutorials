```python
## SEABORN_TUTORIALS


import seaborn as sns
```

```python
df = sns.load_dataset("iris")
```

```python
df.head()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
sns.jointplot(x='petal_width', y='sepal_length', data=df, kind='reg')
```

```python
sns.pairplot(df)
```

```python
sns.distplot(df['sepal_width'],bins=10)
```

```python
sns.countplot('species', data=df)
```

```python
sns.barplot(x='species', y='petal_length', data=df)
```

```python
df.species.value_counts()
```

```python
sns.boxplot(x='species', y='petal_width',data=df)
```

```python
sns.boxplot(x='sepal_length', y='sepal_width', hue='species',data=df)
```

```python
sns.violinplot(x='sepal_length', y='species',data=df)
```
